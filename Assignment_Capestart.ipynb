```python
#Heading preprocessing 
```

```python
stopwords = nltk.corpus.stopwords.words('english')
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
porter_stemmer = PorterStemmer()
```

```python
new_heading = []
for i in data["Heading"]:
    #print(i)
    text = re.sub(r"^http://t.co/[a-zA-Z0-9]*\s"," ",i)
    text=text.lower()
    text="".join([j for j in text if j not in string.punctuation])
    text = re.split(r"\s+", text)
    text= [j for j in text if j not in stopwords]
    text = [porter_stemmer.stem(word) for word in text]
    text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    print(text)
    new_heading.append(text)
    #data["New_Heading"]=[text] * data.shape[0]
    #break
```

```python
data["New_Heading"] = new_heading
```

```python
#Article.Description
```

```python
new_Article_Description = []
for i in data["Article.Description"]:
    #print(i)
    text = re.sub(r"^http://t.co/[a-zA-Z0-9]*\s"," ",i)
    text=text.lower()
    text="".join([j for j in text if j not in string.punctuation])
    text = re.split(r"\s+", text)
    text= [j for j in text if j not in stopwords]
    text = [porter_stemmer.stem(word) for word in text]
    text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    print(text)
    new_Article_Description.append(text)
   
```

```python
data["New_Article_Description"] = new_Article_Description
```

```python
data.head()
```

# For vectorization, use SentenceBERT or other BERT tokenizer.

```python
from bert_embedding import BertEmbedding
bert_embedding = BertEmbedding()
```

```python
Heading_encoding = []
for i in data["New_Heading"]:
    print(i)
    result = bert_embedding(i)
    vector=[]
    for i in range(0,len(result)):
        vector.append(result[i][1])
    Heading_encoding.append(vector)
```

```python
data["Heading_encoding"]=Heading_encoding
```

```python
Article_Description_encoding = []
for i in data["New_Article_Description"]:
    print(i)
    result = bert_embedding(i)
    vector=[]
    for i in range(0,len(result)):
        vector.append(result[i][1])
    Article_Description_encoding.append(vector)
```

```python
data["New_Article_Description_encoding"]=Article_Description_encoding
```

# Feture selection and split data into Train and Test stes 

```python
x = data.iloc[0:, [-1,-2,-7]] 
y = data.iloc[0:, -8].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.20,random_state=1)
```

# Select and train an ML classifier model

```python
from sklearn.tree import DecisionTreeClassifier
reg = DecisionTreeClassifier()
```

```python
#train[0].tolist()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
y_predict = reg.predict(x_test)
```

# Validate and evaluate the Accuracy (Use Cross-validation methods)

```python
from sklearn.model_selection import KFold, cross_val_score
x = data.iloc[0:, [-1,-2,-7]] 
y = data.iloc[0:, -8].values

clf = DecisionTreeClassifier(random_state=42)

k_folds = KFold(n_splits = 5)

scores = cross_val_score(clf, x, y, cv = k_folds)

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```
